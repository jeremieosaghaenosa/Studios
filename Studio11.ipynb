```python
import pandas as pd
sheet = pd.read_excel("StudioResponses.xlsx", "Form Responses 1")
rowLen,columnLen = sheet.shape
sheet
```

```python
studio = {}
studio.clear()
counter = 0
for name in range (rowLen):
    if(sheet["Studio"][name] == "Studio 11"):
        studio[(sheet['WUSTL_Keys'][name])] = counter
        counter = counter + 1
        
        studio[(sheet['WUSTL_Keys.1'][name])] = counter
        counter = counter + 1

        studio[(sheet['WUSTL_Keys.2'][name])] = counter
        counter = counter + 1

        studio[(sheet['WUSTL_Keys.3'][name])] = counter
        counter = counter + 1
        
        studio[(sheet['WUSTL_Keys.4'][name])] = counter
        counter = counter + 1

        studio[(sheet['WUSTL_Keys.5'][name])] = counter
        counter = counter + 1
```
